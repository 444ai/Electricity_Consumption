#### Checking for Missing Values
This command examines our dataset for missing (NA/NaN) values:

**What this code does:**
* `isna()` creates a boolean DataFrame of  `energy_raw` where:
  * `True` indicates a missing value (NA/NaN)
  * `False` indicates a non-missing value
  * Each column is checked individually

**Why this is important:**
* Identifying missing values is crucial because:
  * Missing data can affect statistical calculations
  * Some analysis methods can't handle missing values
  * We might need to:
    * Fill missing values (imputation)
    * Remove rows with missing values
    * Investigate why data is missing

**Note:** 
* Alternative methods:
  * `isnull()` does the same thing as `isna()`
  * `isna().sum()` would give us the count of missing values per column
* This is typically one of the first steps in data cleaning

````python
energy_raw.isna().sum(axis=1)
````

#### Counting Missing Values Per Row
This command counts how many missing values exist in each row:

**What this code does:**
* `isna()`: Creates a boolean DataFrame where:
  * `True` = missing value
  * `False` = non-missing value
* `sum(axis=1)`: Sums across rows (horizontally) where:
  * `axis=1` tells pandas to sum across columns
  * Each `True` counts as 1
  * Each `False` counts as 0
  * Returns a Series where:
    * Index = Time (since that's our index)
    * Value = Number of missing values in that row

**Why this is important:**
* Helps identify:
  * Which timestamps have missing measurements
  * How many measurements are missing at each time
  * Patterns in missing data (e.g., specific time periods with more missing values)
* Useful for:
  * Data quality assessment
  * Deciding whether to drop rows with too many missing values
  * Understanding the completeness of our time series data

````python
energy_raw.isna().sum().sum()
````

#### Getting Total Missing Values in Dataset
This command calculates the total number of missing values across the entire DataFrame:

**What this code does:**
* First `sum()`: Adds up missing values in each column
  * Creates a Series with the count of missing values per column
* Second `sum()`: Adds up all those column totals
  * Gives us a single number representing all missing values in the DataFrame

**Breaking it down:**
1. `isna()` â†’ Creates boolean DataFrame (True for missing values)
2. First `sum()` â†’ Counts missing values per column
3. Second `sum()` â†’ Adds up all column counts

**Why this is important:**
* Provides a quick overview of data completeness
* Helps assess overall data quality
* Useful for:
  * Determining if missing data is a significant issue
  * Deciding on data cleaning strategies
  * Documenting data quality metrics
* Common threshold rules use this total to decide if:
  * Dataset needs extensive cleaning
  * Missing value imputation is necessary
  * Dataset is usable as-is
