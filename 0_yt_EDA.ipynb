#### Analyzing Missing Values by Date

```
Time         Solar    Wind    Battery   Temp
9:00 AM      5        NaN     10       NaN     → 2 missing

```

**Why this is useful:**

````python
# Calculate NaN counts by row and reset index
nan_by_date = energy_raw.isna().sum(axis=1)
nan_by_date = nan_by_date.rename("NA_count").reset_index()

# Create a copy and calculate percentages
percent_nan_by_date = nan_by_date.copy()
percent_nan_by_date["NA_count"] = 100 * percent_nan_by_date["NA_count"] / 172

# Create subplots
fig, ax = plt.subplots(ncols=2, figsize=(20, 5))

# Plot the count of NaNs over time
sns.lineplot(data=nan_by_date, x="Time", y="NA_count", ax=ax[0])

# Plot percentage of NaNs over time (e.g., last 8700 samples)
sns.lineplot(data=percent_nan_by_date.iloc[-8700:], x="Time", y="NA_count", ax=ax[1])

# Add labels
ax[1].set_ylabel("% Devices Down")
plt.show()
````
