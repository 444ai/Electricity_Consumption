#### Checking for Missing Values
This command examines our dataset for missing (NA/NaN) values:

**What this code does:**
* `isna()` creates a boolean DataFrame of  `energy_raw` where:
  * `True` indicates a missing value (NA/NaN)
  * `False` indicates a non-missing value
  * Each column is checked individually

**Why this is important:**
* Identifying missing values is crucial because:
  * Missing data can affect statistical calculations
  * Some analysis methods can't handle missing values
  * We might need to:
    * Fill missing values (imputation)
    * Remove rows with missing values
    * Investigate why data is missing

**Note:** 
* Alternative methods:
  * `isnull()` does the same thing as `isna()`
  * `isna().sum()` would give us the count of missing values per column
* This is typically one of the first steps in data cleaning

#### Counting Missing Values Per Row

  * `axis=1` tells pandas to sum across columns

  * Patterns in missing data (e.g., specific time periods with more missing values)
* Useful for:

```

```

Time         Solar    Wind    Battery

#### Visualizing Missing Data Distribution
This code creates a histogram showing how missing values are distributed across devices:

```

**What the code does:**
* `sns.histplot()`: Creates a histogram using Seaborn library
 * `percent_nan_by_device`: Our data (percentage of missing values for each device)
 * `bins=20`: Divides the data into 20 equal-width bars

**Breaking it down:**
1. Takes our percentages from earlier:
  * Each device has a percentage of missing values
  * Example: Solar: 50%, Wind: 50%, Battery: 25%

2. Creates bars showing:
  * X-axis: Percentage of missing values (0% to 100%)
  * Y-axis: Number of devices with that percentage

**Why this visualization is useful:**
* Quick visual check of data quality

 * Peaks show common missing data percentages

````python
percent_nan_by_device.sort_values(ascending=True)
````

````python
# Which are the devices with highest % of missing values? 
low_val_meters = percent_nan_by_device.sort_values(ascending=False).index[0:5]
low_val_meters
fig, ax = plt.subplots(nrows=5, figsize=(20,10))
for i, column in enumerate(low_val_meters):
    print(column)
    plot1 = sns.lineplot(energy_raw[column], ax = ax[i])
    plot1.set_xlim(energy_raw.index.min(), energy_raw.index.max())
    
````
